### câu 3 :   

Python là ngôn ngữ thông dịch . Vậy ngôn ngữ thông dịch là gì và nó khác gì với ngôn ngưx phiên dịch như C++ ?
- 1 : Python là ngôn ngữ thông dịch (Interpreted language) . Python không biên dịch mã nguồn thành mã máy như C++ hay Java mà nó thực hiện dịch lệnh theo từng dòng và thực thi ngya lập tức . Quá trình này diễn ra từng bước một :     
  - Phân tích cú pháp (Parsing): Trình thông dịch Python phân tích cú pháp của mã nguồn để xác định cấu trúc của chương trình và kiểm tra xem mã có tuân theo cú pháp đúng hay không.

  - Biên dịch thành bytecode: Python sau đó biên dịch mã nguồn thành bytecode, một dạng trung gian của mã máy. Bytecode được lưu trữ trong các tệp .pyc hoặc .pyo.

  - Thực thi: Bytecode sau đó được thực thi bởi máy ảo Python (Python Virtual Machine - PVM) hoặc trình thông dịch Python chính. Máy ảo Python thực thi bytecode dưới dạng các lệnh máy ảo cụ thể.
- **Vậy tại sao python lại được thiết kế như vậy ?**
  - Python được thiết kế như vậy bởi nó hướng đến tính đơn giản và sự linh hoạt . Nó cho phép thưc hiện và kiểm tra từng dòng mã một . giúp phát triển nhanh chóng và dễ dàng kiểm tra lỗi . tuy nhiên điểu này khiến python chậm hơn so với các ngôn ngữ biên dịch như C++ ...

3. các kiễu dữ liêu trong python

- kiểu int : kiểu số nguyên , có thể lưu các số nguyên âm và dương
  - ví dụ : 113 ,-114
- kiểu float : kiểu số thực
  - ví dụ : 5.3, -7.3
- kiểu complex : kiểu số phức
  - ví dụ : z = 2 + 3j thì 2 là phần thức , 3 là phần ảo(j là từ khoá để đánh dấu phân ảo )
  - ví dụ 2: z = complex(2,3) thì 2 là phần thực và 3 là phần ảo .
  - khi xuất ta có thể xuất :    
     - print("Phần thực =" , z.real) ==> phần thực = 2.
     - print("Phần ảo = " , z.imag) ==> Phần ảo = 3 .
- kiểu str : kiểu chuỗi , để trong nháy đôi hoặc nháy đơn
    - ví dụ : "obama"
- kiểu bool : kiểu đúng sai , true- false .
- kiểu dãy ( sequence Types) :     
   - Kiểu danh sách (List) :cho phép bạn lưu trữ nhiều giá trị trong 1 biến
   - kiểu bộ (Tuple) : tương tự như danh sáhc nhưng không thay đổi.
   - kiểu từ điển (dictionary) : kiểu từ điển , cho lưu các gía trị khoá - giá trị ( key- value )
   - kiểu tập hợp (set) : lưu trữ các giá trị không trùng lặp
- kiểu lớp (class) : có thể định nghĩa kiểu dữ liệu riêng của mình bằng cách tạo lớp .
- kiểu hàm (Funtion): như c++ , java.. em không có gì để nói :))
